```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
from scipy.stats import zscore


# Define kitchen appliance-related features
kitchen_appliance_columns = [
    'NUMFRIG', 'SIZRFRI1', 'TYPERFR1', 'AGERFRI1', 'SIZRFRI2', 'TYPERFR2', 'AGERFRI2', 'LOCRFRI2',
    'NUMFREEZ', 'UPRTFRZR', 'SIZFREEZ', 'AGEFRZR',
    'STOVEN', 'STOVENFUEL', 'COOKTUSE', 'OVENUSE', 'STOVE', 'STOVEFUEL', 'OVEN', 'OVENFUEL',
    'MICRO', 'AMTMICRO', 'OUTGRILL', 'OUTGRILLFUEL',
    'TOAST', 'TOASTOVN', 'COFFEE', 'CROCKPOT', 'FOODPROC', 'RICECOOK', 'BLENDER',
    'DISHWASH', 'DWASHUSE', 'DWCYCLE', 'AGEDW'
]

# Select only the relevant columns
df = df[kitchen_appliance_columns]

# Handle missing values (fill with mean)
df.fillna(df.mean(), inplace=True)

### METHOD 1: Isolation Forest for Anomaly Detection ###
iso_forest = IsolationForest(contamination=0.05, random_state=42)
df['Anomaly_IF'] = iso_forest.fit_predict(df)

# Anomalies detected by Isolation Forest
anomalies_if = df[df['Anomaly_IF'] == -1]

### METHOD 2: Z-score for Outlier Detection ###
z_scores = np.abs(df.select_dtypes(include=[np.number]).apply(zscore))  # Apply Z-score
df['Anomaly_Z'] = (z_scores > 3).any(axis=1)  # Mark anomalies where Z-score > 3

# Anomalies detected by Z-score
anomalies_z = df[df['Anomaly_Z']]

### VISUALIZATION ###
plt.figure(figsize=(10, 5))
sns.boxplot(data=df.drop(columns=['Anomaly_IF', 'Anomaly_Z']), orient='h')
plt.title("Boxplot of Kitchen Appliance Features (Detecting Outliers)")
plt.show()

print(f"Total Anomalies detected by Isolation Forest: {len(anomalies_if)}")
print(f"Total Anomalies detected by Z-score: {len(anomalies_z)}")

# Save anomaly records to CSV
anomalies_if.to_csv('anomalies_isolation_forest.csv', index=False)
anomalies_z.to_csv('anomalies_z_score.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Input, Dense
from scipy.stats import zscore



# Define kitchen appliance-related features
kitchen_appliance_columns = [
    'NUMFRIG', 'SIZRFRI1', 'TYPERFR1', 'AGERFRI1', 'SIZRFRI2', 'TYPERFR2', 'AGERFRI2', 'LOCRFRI2',
    'NUMFREEZ', 'UPRTFRZR', 'SIZFREEZ', 'AGEFRZR',
    'STOVEN', 'STOVENFUEL', 'COOKTUSE', 'OVENUSE', 'STOVE', 'STOVEFUEL', 'OVEN', 'OVENFUEL',
    'MICRO', 'AMTMICRO', 'OUTGRILL', 'OUTGRILLFUEL',
    'TOAST', 'TOASTOVN', 'COFFEE', 'CROCKPOT', 'FOODPROC', 'RICECOOK', 'BLENDER',
    'DISHWASH', 'DWASHUSE', 'DWCYCLE', 'AGEDW'
]

# Select only relevant columns
df = df[kitchen_appliance_columns]

# Handle missing values (fill with mean)
df.fillna(df.mean(), inplace=True)

# Standardize data for AutoEncoder and LOF
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

### METHOD 1: Local Outlier Factor (LOF) ###
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
df['Anomaly_LOF'] = lof.fit_predict(df_scaled)
anomalies_lof = df[df['Anomaly_LOF'] == -1]

### METHOD 2: AutoEncoder (Deep Learning) ###
input_dim = df_scaled.shape[1]

# Define AutoEncoder model
autoencoder = Sequential([
    Dense(16, activation='relu', input_shape=(input_dim,)),
    Dense(8, activation='relu'),
    Dense(16, activation='relu'),
    Dense(input_dim, activation='linear')
])

autoencoder.compile(optimizer='adam', loss='mse')

# Train AutoEncoder
autoencoder.fit(df_scaled, df_scaled, epochs=50, batch_size=16, shuffle=True, verbose=0)

# Reconstruction error (Mean Squared Error)
reconstructions = autoencoder.predict(df_scaled)
mse = np.mean(np.power(df_scaled - reconstructions, 2), axis=1)

# Set threshold for anomaly detection
threshold = np.percentile(mse, 95)
df['Anomaly_AE'] = mse > threshold
anomalies_ae = df[df['Anomaly_AE']]

### VISUALIZATION ###
plt.figure(figsize=(10, 5))
sns.boxplot(data=df.drop(columns=['Anomaly_LOF', 'Anomaly_AE']), orient='h')
plt.title("Boxplot of Kitchen Appliance Features (Detecting Outliers)")
plt.show()

# Print results
print(f"Total Anomalies detected by LOF: {len(anomalies_lof)}")
print(f"Total Anomalies detected by AutoEncoder: {len(anomalies_ae)}")

# Save anomaly records to CSV
anomalies_lof.to_csv('anomalies_lof.csv', index=False)
anomalies_ae.to_csv('anomalies_autoencoder.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.preprocessing import StandardScaler


# Select relevant features
selected_columns = ['DISHWASH', 'HEATHOME', 'AIRCOND']
df_selected = df[selected_columns].dropna()  # Remove NaNs

# Standardize the data for better anomaly detection
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df_selected), columns=selected_columns)

# Isolation Forest
iso_forest = IsolationForest(contamination=0.05, random_state=42)
df_selected['Anomaly_IF'] = iso_forest.fit_predict(df_scaled)

# Local Outlier Factor (LOF)
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
df_selected['Anomaly_LOF'] = lof.fit_predict(df_scaled)

# Convert anomaly labels (-1 = anomaly, 1 = normal) to binary (0 = normal, 1 = anomaly)
df_selected['Anomaly_IF'] = df_selected['Anomaly_IF'].apply(lambda x: 1 if x == -1 else 0)
df_selected['Anomaly_LOF'] = df_selected['Anomaly_LOF'].apply(lambda x: 1 if x == -1 else 0)

# 2D Visualization using Pairplot
sns.pairplot(df_selected, hue='Anomaly_IF', palette={0: 'blue', 1: 'red'})
plt.suptitle("Pairplot - Isolation Forest Anomalies", y=1.02)
plt.show()

sns.pairplot(df_selected, hue='Anomaly_LOF', palette={0: 'blue', 1: 'red'})
plt.suptitle("Pairplot - LOF Anomalies", y=1.02)
plt.show()

# 3D Visualization
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with anomalies
colors = df_selected['Anomaly_IF'].map({0: 'blue', 1: 'red'})
ax.scatter(df_selected['DISHWASH'], df_selected['HEATHOME'], df_selected['AIRCOND'], c=colors, marker='o', alpha=0.8)

# Labels and title
ax.set_xlabel('DISHWASH')
ax.set_ylabel('HEATHOME')
ax.set_zlabel('AIRCOND')
ax.set_title("3D Scatter Plot - Isolation Forest Anomalies")

plt.show()
```

```python
# Count anomalies detected by each model
anomaly_if_count = df_selected['Anomaly_IF'].sum()
anomaly_lof_count = df_selected['Anomaly_LOF'].sum()

# Total number of data points
total_samples = len(df_selected)

print(f"Isolation Forest detected {anomaly_if_count} anomalies out of {total_samples} samples ({(anomaly_if_count/total_samples)*100:.2f}%)")
print(f"Local Outlier Factor detected {anomaly_lof_count} anomalies out of {total_samples} samples ({(anomaly_lof_count/total_samples)*100:.2f}%)")
```
